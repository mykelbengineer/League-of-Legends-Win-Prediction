```python
# https://www.kaggle.com/datasets/vivovinco/league-of-legends-stats-s13/
stats = pd.read_csv('League of Legends Champion Stats 13.1.csv', sep=';')
stats = stats[['Name', 'Tier', 'Role']]
```

```python
def normit(a):
    s = ''
    for i in a:
        if i.isalpha() or i.isspace():
            s+=i
    
    s = s.replace('Dr Mundo', 'DrMundo')       # This is the only inconsistency between stats and pre
    return s
```

```python
stats['Name']=stats['Name'].apply(normit)
```

```python
# mapping Tier and Role to the name on each positions below

positions = ['JUNGLE1', 'JUNGLE2', 'TOP1', 'TOP2', 'MID1', 'SUPPORT2', 'MID2', 'SUPPORT1', 'CARRY1', 'CARRY2']

pre2 = pre.copy()
for i in positions:
    
    # rename CARRY as ADC (Attack Damage Carry)
    posit = i.replace('CARRY','ADC')                         
    
    # remove the number that represent the teams. e.g. 'TOP1' -> 'TOP'
    stats_filtered = stats[stats['Role'] == posit[:-1]]
    
    # develop dict for mapping the tier
    name_to_tier = dict(zip(stats_filtered['Name'],stats_filtered['Tier']))    
    
    # map the tier to each character for their role
    pre2[i+'Tier'] = pre2[i].map(lambda x: name_to_tier.get(x, None))    
    
display(pre2)
```

```python

r = ['MID', 'SUPPORT', 'JUNGLE', 'ADC', 'TOP']         # positions on stats as roles
r2 = ['MID', 'SUPPORT', 'JUNGLE', 'CARRY', 'TOP']      # positions on pre as column names

for position, col_name in zip (r,r2):
    for i in range(2):
        m1 = stats[stats['Role'] == position][['Name', 'Tier']].rename(columns={'Name': col_name+str(i+1)})
        pre = pre.merge(m1, on=col_name+str(i+1), how='left').rename(columns={'Tier': col_name + str(i+1) + 'Tier'})

display(pre)
```
