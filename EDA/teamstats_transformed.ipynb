```python
# transform the teamstat table 
# original: there are 2 rows per matchID: 1 per team, namely 100 and 200.
# transformation: (1) change to 1 row per matchID, (2) remove unused columns, (3) rename columns so each team
#                 has the own header with team# at the suffix.
# For example, the original table has column: matchid, teamid, firsttower, firstinhib,...
#        after transformed -> column: matchid, firsttower1, firsttower2, firstinhib1, firstinhib2, ... 
#        notice that 'teamid' has been dropped

import pandas as pd

d = pd.read_csv("teamstats.csv")
d = d.drop(columns=['firstblood','firstdragon','harrykills'])  # drop unused columns
team1 = d[d['teamid']==100]       # split the tables by teamid. teamid is either '100' or '200'
team1.columns = [x+"1" if x != "matchid" else x for x in d.columns]  # rename columns for team1 by adding '1' to suffix
team2 = d[d['teamid']==200]
team2.columns = [x+"2" if x != "matchid" else x for x in d.columns]
teamstats_transformed = team1.merge(team2, on="matchid").drop(columns=["teamid1",'teamid2']) 
display(teamstats_transformed)
```
